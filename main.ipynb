### AIM: Mall Customer Segmentation using Clustering algorithms

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, davies_bouldin_score
import time
import skfuzzy as fuzz
```

```python
data = pd.read_csv('Mall_Customers.csv')
X = data[['Annual Income (k$)', 'Spending Score (1-100)']].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

plt.figure(figsize=(10, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c='blue', alpha=0.5)
plt.title('Unclustered Mall Customer Data')
plt.xlabel('Standardized Annual Income (k$)')
plt.ylabel('Standardized Spending Score (1-100)')
plt.show()
```

```python
def kmeans_clustering(data, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, max_iter=1, n_init=1, random_state=42)
    centroids_history = []
    iteration = 0
    previous_centroids = None
    start_time = time.time()
    
    print("\nK-means Clustering:")
    while True:
        iteration += 1
        kmeans.fit(data)
        centroids = scaler.inverse_transform(kmeans.cluster_centers_)
        centroids_history.append(centroids)
        
        print(f"\nIteration {iteration} Centroids:")
        for j, centroid in enumerate(centroids):
            print(f"Cluster {j}: Income = {centroid[0]:.2f}k$, Spending Score = {centroid[1]:.2f}")
        
        plt.figure(figsize=(10, 6))
        plt.scatter(data[:, 0], data[:, 1], c=kmeans.labels_, cmap='viridis', alpha=0.5)
        plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
                   c='red', marker='x', s=200, linewidths=3, label='Centroids')
        plt.title(f'K-means Clustering - Iteration {iteration}')
        plt.xlabel('Standardized Annual Income (k$)')
        plt.ylabel('Standardized Spending Score (1-100)')
        plt.legend()
        plt.show()
        
        if previous_centroids is not None and np.allclose(previous_centroids, centroids, rtol=1e-4):
            break
        previous_centroids = centroids.copy()
    
    end_time = time.time()
    kmeans_time = end_time - start_time
    
    print("\nFinal K-means Centroids:")
    for j, centroid in enumerate(centroids):
        print(f"Cluster {j}: Income = {centroid[0]:.2f}k$, Spending Score = {centroid[1]:.2f}")
    
    silhouette = silhouette_score(data, kmeans.labels_)
    db_index = davies_bouldin_score(data, kmeans.labels_)
    
    return kmeans, centroids_history, kmeans_time, silhouette, db_index, iteration

def fuzzy_cmeans_clustering(data, n_clusters):
    if fuzz is None:
        print("\nFuzzy C-means Clustering skipped: skfuzzy module not installed")
        return None, None, None, None, None, None
    
    print("\nFuzzy C-means Clustering:")
    data_reshape = data.T
    centroids_history = []
    iteration = 0
    previous_u = None
    start_time = time.time()
    
    cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(
        data_reshape, c=n_clusters, m=2, error=0.005, maxiter=1, init=None, seed=42
    )
    
    while True:
        iteration += 1
        centroids = scaler.inverse_transform(cntr)
        centroids_history.append(centroids)
        
        print(f"\nIteration {iteration} Centroids:")
        for j, centroid in enumerate(centroids):
            print(f"Cluster {j}: Income = {centroid[0]:.2f}k$, Spending Score = {centroid[1]:.2f}")
        
        cluster_membership = np.argmax(u, axis=0)
        
        plt.figure(figsize=(10, 6))
        plt.scatter(data[:, 0], data[:, 1], c=cluster_membership, cmap='viridis', alpha=0.5)
        plt.scatter(cntr[:, 0], cntr[:, 1], c='red', marker='x', s=200, linewidths=3, label='Centroids')
        plt.title(f'Fuzzy C-means Clustering - Iteration {iteration}')
        plt.xlabel('Standardized Annual Income (k$)')
        plt.ylabel('Standardized Spending Score (1-100)')
        plt.legend()
        plt.show()
        
        if previous_u is not None and np.allclose(previous_u, u, rtol=1e-4):
            break
        
        previous_u = u.copy()
        um = u ** 2
        cntr = (um @ data) / um.sum(axis=1)[:, None]
        d = np.sum((data[:, None, :] - cntr[None, :, :]) ** 2, axis=2)
        u = 1 / ((d / d.min(axis=1)[:, None]) + 1e-10)
        u = u.T
    
    end_time = time.time()
    fcm_time = end_time - start_time
    
    print("\nFinal Fuzzy C-means Centroids:")
    for j, centroid in enumerate(centroids):
        print(f"Cluster {j}: Income = {centroid[0]:.2f}k$, Spending Score = {centroid[1]:.2f}")
    
    silhouette = silhouette_score(data, cluster_membership)
    db_index = davies_bouldin_score(data, cluster_membership)
    
    return cntr, u, fcm_time, silhouette, db_index, iteration
```

```python
n_clusters = int(input("Enter the number of clusters: "))
print("Running clustering algorithms on Mall Customer Dataset:")
kmeans_model, kmeans_history, kmeans_time, kmeans_silhouette, kmeans_db, kmeans_iters = kmeans_clustering(X_scaled, n_clusters)
fcm_centroids, fcm_membership, fcm_time, fcm_silhouette, fcm_db, fcm_iters = fuzzy_cmeans_clustering(X_scaled, n_clusters)
```
